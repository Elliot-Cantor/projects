```python
def mean_normalization(label):
    data = df[label].values
    normal = []
    for i in range(len(data)):
        normed = (data[i] - np.average(data)) / (np.max(data) - np.min(data))
        normal.append(normed)
    return np.array(normal)

def min_normalization(label):
    data = df[label].values
    normal = []
    for i in range(len(data)):
        normed = (data[i] - np.min(data)) / (np.max(data) - np.min(data))
        normal.append(normed)
    return np.array(normal)   
```

A large advantage of using an optimize function over neural networks to model the stellar lables is that you have much more control over what form you want the model to take, whereas in machine learning you can pick different activation functions and change a few other variables, but you still have much less control over the model. It is also very challanging to predict the effectiveness of a neural network as seemingly small changes can have wide ranging effects on the results of the network. While in an optimizer you have a much better sense on how your guess labels for your model can impact your best solution that the optimizer returns, and how changes to your log likelihood function can change the effectiveness of the model. A large advantage machine learning has over an optimizing function based model is that you can feed it huge amounts of data and have it create models in drastically less time than it would take an optimizer, which is very true for this data set, as my label model takes about 10 minutes to run after spending an hour making my coefficient matrix, while my machine learning model only takes about 5 minutes if I bump up the epochs to a large value.
